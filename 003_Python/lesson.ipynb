WHILE


```python
import time #bulit in funk lar print gibi direk kullanabiliyoruz fakat bazi modulleri kodumuzda ihtiyac oldugunda kullanmak icin import etmemiz gerekiyor bunlar yine python ile geliyor fakat import olmadan kullanilamiyor  time modulu beklme yapmamizi sagliyor ve sleep metodu ile saniye cinsinden bekleme yapiyor
num = 5
while num > 0:
    print(f'countdown {num}') #print satirinin yerine göre önce yazar sonra islemi yapar f kullanarak () icini degisken sekilde yazabiliyoruz 
    time.sleep(1)
    num -= 1

print("Fire!!") # while block unun disina print yazarak döngu bitince yaz demis oluyoruz 
```

```python
pw = '' #bos '' olrak degiskeni önce tanimliyoruz 
giris_hakki = 3 #kullaniciya giris icin sinir koyuyoruz

while pw != 'secret' and giris_hakki > 0:  #pw yi dogru girene kadar döngu devam edecek and ile 2.condition u birlestirebiliriz
    pw = input('Sifreyi giriniz: ') #input fonk kullaniciya girdi yaptirir
    giris_hakki -= 1
    print(f'{giris_hakki} hakkiniz kaldi') # döngu calistikca user a feedback veriyoruz

print("lütfen yönetici ile iletişime geçiniz") # döngu bitince user a bilgi veriyor 

#dogru girilme durumu eksik
```

```python
name = "yasin"
for i in name:
    print(i)
```

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

```python
for num in range(3, 50, 10):
    print(num)
```

```python
adjs = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for adj in adjs:
    for fruit in fruits:

        print(adj, fruit)
```

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:

    if fruit == "banana":
        break
    print(fruit)
```

```python
num = 1

while num < 10:
    print(num)
    num += 1
    if num == 5:
        break
```

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:

    if fruit == "apple":
        continue
    print(fruit)
```
