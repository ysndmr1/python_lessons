### Keyword Arguments


```python
def salute_to_name(first_name,last_name):
    return f'Hello {first_name} {last_name}'
```

```python
salute_to_name(first_name='Anne', last_name='Mary')
```

### **kwargs


```python
def my_kids(kid1, *args,**kwargs):

    print(f'Hello {kid1}')

    for arg in args:
        print(f'Hello {arg} arbitrary args')
    for kid in kwargs.values():
        print(f'Hello {kid} kwargs')
```

```python
my_kids('Mary', 'Tommy',kid2='Asya',kid3='Eva')
```
