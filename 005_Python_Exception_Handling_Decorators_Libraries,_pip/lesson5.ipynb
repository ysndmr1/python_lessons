```python
num1= 5
num2=0
# olusacak bir hata ile ilgili exceptional handling yaziyoruz
# try da yapacagimiz islemi altina da gerceklesecek exception i yaziyoruz 
# try satiri ni yapar calismassa except satirini calistirir ve programin cökmesini de engellemis oluyor 
# return ile yazmak istersek bir fonksiyon ile beraber kullanmamiz gerekiyor 
# eger except blogunda gerceklescek hatanin ne oldugunu biliyorsak o hatanin adini da yazabiliyoruz ZeroDivisionError gibi 
# eger belirttigimiz hata disinda baska bir hata olusursa tekrardan kod calismayabilir o yuzden diger olusacak hatayi da yazmamiz gerekir 
# eger sadece except yazarsak o olusacak butun hatalari kapsar
# eger mesajin yaninda hangi hata type inin da oldgunu görmek istersek error ismini alise kullanarak as e seklinde yapip onu da print ile yazdirabiliriz yada direk olarak print icinde hata class ini yazabiliriz print ('ZeroDivisionError') seklinde 
try:
   num1/num2
except ZeroDivisionError as e:
   print(f'It is not possible to divide by {num2}')
   print(e)
except TypeError:
   print('there is an error in data type')  
except:
   print('This is a general error message')      
```

```python
# örnegi try except blogu icinde yazmak istersek 

def calc_bmi(height,weight):
   
   try:
    bmi=round(weight/(height ** 2))
    if bmi > 30:
        print(f'Your BMI = {bmi}.You are obese')
    elif bmi >25:
        print(f'Your BMI = {bmi}.You are overweight')
    elif bmi <=25 and bmi>18.5 :
        print(f'Your BMI = {bmi}.You are normal')
    else:
        print('You are thin') 
   except TypeError:
    print('Wrong data type')  
   except:
      print('Another error type')       
                    
calc_bmi(1.85,'sas')           


```

<h3>Decorators

```python
#decorators special kind of function bir fonksiyona istedigimiz bir özellik eklemek icin kullanabiliriz 
#fonk parametre olarak diger fonk u alir 
# Decorators are really useful when you want to apply a common behavior or feature to many functions without duplicating code.
# dahil ettigimiz fonk öncesine veya sonrasina yada sadece önune veya arkasina ekleme yapabliyoruz 

def my_decorator(func):
    def wrapper():
        # Do something before
        result = func()  # or only func()
        # Do something after
        return result  # or erase this line, with only func() above
    return wrapper

@my_decorator
def func():
    pass
```

```python
def my_decorator(func):
    def wrapper():
        # Do something before
        print('--------------------')
        result = func()  # or only func()
        print(result)
        print('--------------------')
        # Do something after
        #return result  # or erase this line, with only func() above
    return wrapper

@my_decorator
def func(): #decorator in icine yazdigimiz fonk burda tanimliyoruz sadece hello yaziyor, decoratorda yazarken onune veya arkasna ekleme yapabiliyoruz 
    return 'Hello'
```

```python
import time

def gecen_zaman(func):
    def wrapper():
        # Do something before
        t1 = time.time() # ilk zamani t1 ile aliyor 
        func()  # or only func()
        # Do something after
        t2 = time.time() # fonks dan sonra zamani bir daha aliyoruz arada gececek olan zaman icin 
        return  f'gecen zaman: {t2 - t1}' # or erase this line, with only func() above
    return wrapper

@gecen_zaman
def big_sum():
    num_list = []
    for num in (range(0, 10000)):
        num_list.append(num)
    print(f'Big sum: {sum(num_list)}')
```

```python
big_sum()
```
